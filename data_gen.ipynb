```python
columns = ['industry', 'store', 'age', 'gender', 'revisit', 'emotion', 'time_spent']
main_data = {column:list() for column in columns}
for industry in data_dict_1:
    entries = data_dict_1[industry][0]
    for column in data_dict_1[industry][1]:
        col_dic = data_dict_1[industry][1][column]
        data = []
        if column=='age' or column=='time_spent':
            for start,end in col_dic:
                countval = int((col_dic[(start,end)]/100.0)*entries)
                data.extend([random.randint(start,end) for _ in range(countval)])
        if column=='gender' or column=='emotion':
            for key,count in col_dic.items():
                countval = int((count/100.0)*entries)
                data.extend([key for _ in range(countval)])
            for _ in range(100):
                random.shuffle(data)
        if column=='revisit':
            countval = int((col_dic/100.0)*entries)
            data.extend([1 for _ in range(countval)])
            data.extend([0 for _ in range(entries-countval)])
            for _ in range(100):
                random.shuffle(data)
        if(len(data)!=0):
            data.extend([random.choice(data) for _ in range(entries-len(data))])
        main_data[column].extend(data)
    main_data['industry'].extend([industry for _ in range(entries)])
    main_data['store'].extend([1 for _ in range(entries)])
```
